```python
def LoadFeatures(path):

    df_train = pd.read_csv(path+'train.csv')
    X_train = df_train.drop('Label', axis=1) 
    y_train = df_train['Label']  

    df_test = pd.read_csv(path+'test.csv')
    X_test = df_test.drop('Label', axis=1) 
    y_test = df_test['Label']  

    df_test_ev = pd.read_csv(path+'test_evento.csv')
    #df_test_ev = df_test_ev.sample(frac=1).reset_index(drop=True)
    X_test_ev = df_test_ev.drop('Label', axis=1) 
    y_test_ev = df_test_ev['Label']  

    #return X_train, y_train, X_test[:1000], y_test[:1000], X_test_ev[:1000], y_test_ev[:1000]
    return X_train, y_train, X_test, y_test, X_test_ev, y_test_ev
```

```python
from sklearn.feature_selection import mutual_info_classif
import pandas as pd

path = 'dataset/eventos_dataset_test/exp8/'
X_train, y_train, X_test, y_test, X_test_ev, y_test_ev = LoadFeatures(path)

ganho_informacao = mutual_info_classif(X_train, y_train)

df_ganho = pd.DataFrame({
    'Feature': X_train.columns,
    'Ganho de Informação': ganho_informacao
})

df_ganho_ordenado = df_ganho.sort_values(by='Ganho de Informação', ascending=False)
print("Dataset - Train")
print(df_ganho_ordenado)
print("\n")

###  Dataset Test

ganho_informacao = mutual_info_classif(X_test, y_test)

df_ganho = pd.DataFrame({
    'Feature': X_train.columns,
    'Ganho de Informação': ganho_informacao
})

df_ganho_ordenado = df_ganho.sort_values(by='Ganho de Informação', ascending=False)
print("Dataset - Test")
print(df_ganho_ordenado)
print("\n")

### Dataset Test Event

ganho_informacao = mutual_info_classif(X_test_ev, y_test_ev)

df_ganho = pd.DataFrame({
    'Feature': X_train.columns,
    'Ganho de Informação': ganho_informacao
})

df_ganho_ordenado = df_ganho.sort_values(by='Ganho de Informação', ascending=False)
print("Dataset - Test Event")
print(df_ganho_ordenado)
print("\n")
```

```python
def minShape(lista):
    menor = min(lista) 
    return menor
```

```python
def undersamplingDataset(lista, min):

    listUnderSampling = []
    for df_train in lista:
        # Embaralhando as instâncias
        train_data = df_train.sample(frac=1).reset_index(drop=True)

        # Undersampling => realizando a poda 
        train_data = train_data[:min]
        
        listUnderSampling.append(train_data)
    
    return listUnderSampling
```

```python
path = "dataset/all3"
path_out = 'dataset/eventos_dataset_ok'

for indice, lista in enumerate(listas_train):

    indice = indice + 1
    listaDF = []
    listShape = []
    for evento in lista:
        print(evento)
        df = pd.read_csv(path + '/'+ evento +'/train.csv')
        listaDF.append(df)
        listShape.append(df.shape[0])
        
    print(listShape)
    menor = minShape(listShape) # pega a qtde de instâncias do menor dataset => para o undersampling
    listaTrain = undersamplingDataset(listaDF, menor) 

    train_data = pd.concat(listaTrain, ignore_index=True)
    train_data = train_data.sample(frac=1).reset_index(drop=True)

    os.mkdir(path_out + '/exp' + str(indice))

    train_data.to_csv(path_out + '/exp'+ str(indice) +'/train.csv', index=False)

    print(lista[0]+"\n")


for indice, lista in enumerate(listas_test):
    indice = indice + 1
    listaDF = []
    for evento in lista:
        df = pd.read_csv(path + '/'+ evento +'/test.csv')
        listaDF.append(df)
 
    test_data = pd.concat(listaDF, ignore_index=True)
    test_data = test_data.sample(frac=0.01).reset_index(drop=True)
    
    test_data.to_csv(path_out + '/exp'+ str(indice) +'/test.csv', index=False)

    print(lista[0]+"\n")


for indice, lista in enumerate(listas_test):
    indice = indice + 1
    listaDF = []
    for evento in lista:
        df = pd.read_csv(path + '/'+ evento +'/val.csv')
        listaDF.append(df)
 
    test_data = pd.concat(listaDF, ignore_index=True)
    test_data = test_data.sample(frac=0.01).reset_index(drop=True)
    
    test_data.to_csv(path_out + '/exp'+ str(indice) +'/val.csv', index=False)

    print(lista[0]+"\n")

for indice, lista in enumerate(listas_train): # => considerar os mesmos eventos de treino
    indice = indice + 1
    listaDF = []
    for evento in lista:
        df = pd.read_csv(path + '/'+ evento +'/test.csv')
        listaDF.append(df)
 
    test_data = pd.concat(listaDF, ignore_index=True)
    test_data = test_data.sample(frac=0.01).reset_index(drop=True)
    
    test_data.to_csv(path_out + '/exp'+ str(indice) +'/test_evento.csv', index=False)

    print(lista[0]+"\n")
```
